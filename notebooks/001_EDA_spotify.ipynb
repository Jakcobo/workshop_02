### Correlation Heatmap of Numeric Features

The following code selects only the numeric features from the dataset and computes their correlation matrix. A heatmap is then generated to visualize the correlations between these features. This visualization helps identify strong positive or negative correlations, which can provide insights into the relationships between variables in the dataset.

```python
numeric_df = spotify_df.select_dtypes(include=[np.number])
corr_matrix = numeric_df.corr()

plt.figure(figsize=(9,6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```


### Explicit Content Distribution

The output shows that the majority of tracks in the dataset are non-explicit, with 104,253 tracks marked as non-explicit and only 9,747 tracks marked as explicit. This indicates that explicit content constitutes a smaller portion of the dataset, which may reflect the general trend in music production or the curation of this specific dataset.

```python
explicit_counts = spotify_df['explicit'].value_counts()
explicit_counts
```

### Analysis of Track Genres in the Spotify Dataset

The code analyzes the distribution of track genres in the Spotify dataset by counting the occurrences of each genre. The output shows that all 114 unique genres have an equal count of 1,000 tracks, indicating a balanced dataset in terms of genre representation.


```python
genre_counts = spotify_df['track_genre'].value_counts()
genre_counts.head(114)
```

### Distribution of Explicit Column

The following visualization presents a count plot of the `explicit` column, which indicates whether a track contains explicit content. This plot helps to understand the distribution of explicit and non-explicit tracks in the dataset.

```python
plt.figure(figsize=(6,3))
sns.countplot(x='explicit', data=spotify_df)
plt.title("Distribution of Explicit Column")
plt.xlabel("Explicit")
plt.ylabel("Count")
plt.show()
```

### Spotify Popularity Distribution
This section visualizes the distribution of the 'popularity' variable.

```python
plt.figure(figsize=(8, 4))
sns.histplot(spotify_df['popularity'], bins=20, kde=True)
plt.title("Spotify Popularity Distribution")
plt.xlabel("Popularity")
plt.ylabel("Frequency")
plt.show()
```

### Pairplot of Selected Features

The following code generates a pairplot for the selected features: `popularity`, `danceability`, `energy`, `loudness`, and `tempo`. A pairplot is a grid of scatterplots that visualizes the relationships between multiple variables in a dataset. It helps identify trends, correlations, and potential outliers among the selected features.



```python
selected_features = ['popularity', 'danceability', 'energy', 'loudness', 'tempo']
sns.pairplot(spotify_df[selected_features])
plt.show()
```

---
### Boxplot of Song Duration

The following visualization presents a boxplot of the `duration_ms` column, which represents the duration of songs in milliseconds. This plot helps identify the distribution, central tendency, and potential outliers in the song durations.


```python
plt.figure(figsize=(6,3))
sns.boxplot(x=spotify_df['duration_ms'])
plt.title("Boxplot of Song Duration (ms)")
plt.xlabel("Duration (ms)")
plt.show()
```

--- 

### Scatter Plot Analysis: Energy vs Loudness

The scatter plot shows a positive correlation between energy and loudness. As the energy of a track increases, its loudness tends to increase as well. However, there is significant variability at lower energy levels, indicating that tracks with low energy can have a wide range of loudness values.

```python
plt.figure(figsize=(6, 4))  
sns.scatterplot(x='energy', y='loudness', data=spotify_df, alpha=0.5)
plt.title("Scatter Plot of Energy vs Loudness")
plt.xlabel("Energy")
plt.ylabel("Loudness")
plt.grid(True)
plt.show()
```
