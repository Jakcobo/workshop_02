# Exploratory Data Analysis – Spotify

In this notebook we will perform the exploratory data analysis based on the Spotify csv database located in the __/data__ folder.

### Enviroment

```python
import sys
import os


root_path = os.path.abspath(os.path.join('..'))
if root_path not in sys.path:
    sys.path.append(root_path)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

### Connection to data

```python
spotify_df = pd.read_csv("../data/dataset_spotify.csv")
```

```python
spotify_df.head()
```

```python
print(spotify_df.info())
```

```python
spotify_df.describe(include='all')
```

```python
columns_info = pd.DataFrame({
    'Column': spotify_df.columns,
    'Data Type': spotify_df.dtypes,
    'Missing Values': spotify_df.isnull().sum(),
    'Unique Values': [spotify_df[col].nunique() for col in spotify_df.columns]
})
columns_info
```

```python
duplicate_count = spotify_df.duplicated().sum()
duplicate_count 
```

```python
missing_percentage = (spotify_df.isnull().mean())*100 
missing_percentage
```

Only rows with nulls data in the columns: __'artists', 'album_name', 'track_name'__

```python
null_cases = spotify_df[spotify_df[['artists', 'album_name', 'track_name']].isnull().any(axis=1)]
null_cases
```

```python
dataset_shape = spotify_df.shape
dataset_size_in_bytes = spotify_df.memory_usage(deep=True).sum()
dataset_shape, dataset_size_in_bytes
```

```python
numeric_df = spotify_df.select_dtypes(include=[np.number])
corr_matrix = numeric_df.corr()

plt.figure(figsize=(9,6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
explicit_counts = spotify_df['explicit'].value_counts()
explicit_counts
```

```python
genre_counts = spotify_df['track_genre'].value_counts()
genre_counts.head(114)
```

```python
plt.figure(figsize=(6,3))
sns.countplot(x='explicit', data=spotify_df)
plt.title("Distribution of Explicit Column")
plt.xlabel("Explicit")
plt.ylabel("Count")
plt.show()
```

### Spotify Popularity Distribution
This section visualizes the distribution of the 'popularity' variable.

```python
plt.figure(figsize=(8, 4))
sns.histplot(spotify_df['popularity'], bins=20, kde=True)
plt.title("Spotify Popularity Distribution")
plt.xlabel("Popularity")
plt.ylabel("Frequency")
plt.show()
```

```python
selected_features = ['popularity', 'danceability', 'energy', 'loudness', 'tempo']
sns.pairplot(spotify_df[selected_features], height=1.5)
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.boxplot(x=spotify_df['duration_ms'])
plt.title("Boxplot of Song Duration (ms)")
plt.xlabel("Duration (ms)")
plt.show()
```
